### Random Noise added and we want our Network to Learn This 

```python
x_data = [0, 1, 2, 3, 4, 5,6]
y_data = []

# Randomly initialized 
slope = -0.12
b =1.22

for i in x_data:
    y_data.append(slope*i+b)

%matplotlib inline
plt.plot(x_data,y_data)
```

### Import Library 

```python
import tensorflow as tf
import numpy as np
```

### Create a Variable for Weight  and Bias 

```python
W = tf.Variable(0.3, name="Tf_W")
B = tf.Variable(-0.3, name="Tf_B")
```

### Create a place Holder for X and Y

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

### Create a Neural Network

```python
S = W * X + B
A = tf.sigmoid(S)
error = tf.square(-1*(Y- A))
loss = tf.reduce_sum(error)
```

### Create a Optimizer 

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
```

### Create a Global variable Initializer

```python
init = tf.global_variables_initializer()
```

### Run the Code 

```python
sess = tf.Session()
sess.run(init)
for i in range(100):
    sess.run(train, {X:[0, 1, 2, 3, 4, 5,6], Y:[1, 2, 3, 4, 5, 6,7]})
print(sess.run([W,B]))
```
